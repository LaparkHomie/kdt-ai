```python
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips["tip_pct"] = tips["tip"] / (tips["total_bill"] - tips["tip"])
tips.head()
```

```python
sns.set_style("white") # darkgrid, whitegrid, ticks, dark, white
sns.set_context('notebook') # paper, notebook. talk, poster
sns.set_palette('bright') # deep, muted, pastel, dark, bright, colorblind
sns.barplot(x="tip_pct", y="day", hue="time", data=tips, orient="h")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris") # iris, titanic, tips, filghts
x = iris.petal_length.values
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
sns.rugplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Rug Plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris") # iris, titanic, tips, flights
x = iris.petal_length.values
plt.rcParams["font.family"] = 'Malgun Gothic'
sns.kdeplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Kernel Density Plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris") # iris, titanic, tips, flights
x = iris.petal_length.values
plt.rcParams["font.family"] = 'Malgun Gothic'
sns.distplot(x, kde=True, rug=True)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Dist Plot")
plt.show()
```

```python
import seaborn as sns

iris = sns.load_dataset("iris")
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind='hex', color="g") # scatter, kde, hex
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.02)
plt.show()
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips, hue="sex")
plt.title("요일 별, 전체 팁")
plt.show()
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
sns.boxplot(y="day", x="total_bill", data=tips, orient='h')
plt.title("요일 별 전체 팁의 Box Plot")
plt.show()
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("요일별 전체 팁의 Violin Plot")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, palette="muted")
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.title("요일 별 전체 팁의 Strip Plot")
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.title("요일 별 전체 팁의 Swarm Plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib as mpl

mpl.rcParams['axes.unicode_minus'] = False
titanic = sns.load_dataset("titanic")
data = titanic[titanic.survived.notnull()]
sns.catplot(x="age", y="sex", hue="survived", row="class", data=data, kind="bar", height=2, aspect=4)
plt.title("각 클래스의 성별 생존자 수의 Catplot")
plt.show()
```

```python
sns.catplot(x='who', y='survived', col='pclass', row='embarked', kind='bar', height=2.5, data=titanic)
plt.show()
```

```python
import seaborn as sns
import numpy as np

tips = sns.load_dataset("tips")
plt.title("Boxplot과 Strip Plot로 표현한, 요일 별 팁")
sns.boxplot(x="tip", y="day", data=tips, whis=np.inf)
sns.stripplot(x="tip", y="day", data=tips, jitter=True, color="0.4")
plt.show()
```
