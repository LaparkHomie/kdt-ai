```python
def n_square(x):
    return x ** 2
```

```python
def n_square(x, n):
    return x ** n;
```

```python
df4 = df[["Age", "SibSp", "Fare"]].apply(n_square, n=2)
df4
```

 ### 객체 연산

```python
data = [[1,10,100],[2,20,200],[3,30,300]]
col = ['col1', 'col2', 'col3']
row = ['row1', 'row2', 'row3']
df = pd.DataFrame(data=data, index=row, columns=col)
df
```

```python
result = df.add(1) # sub() mul() div() mod()
result
```

```python
result = df + 1
result
```

```python
data2 = [[3], [4], [5]]
df2 = pd.DataFrame(data=data2, index=['row1', 'row2', 'row3'], columns=['col1'])
df2
```

```python
result = df.add(df2)
result
```

```python
result = df.add(df2, fill_value=0) # df2에 없는 컬럼에 0을 채워서 연산
result
```
