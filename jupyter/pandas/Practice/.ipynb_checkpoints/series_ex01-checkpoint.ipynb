```python
data = [1000, 2000, 3000]
index = ['메로나', '구구콘', '하겐다즈']
s = Series(data=data, index=index)
print(s)
s2 = s.reindex(["메로나", "비비빅", "구구콘"])
print(s2)
```

```python
s2 = s2.fillna(0)
print(s2)
```

```python
from pandas import Series

data = [1000, 2000, 3000]
index = ["메로나", "구구콘", "하겐다즈"]
s = Series(data=data, index=index)
s
```

```python
print(s.iloc[0:2])
```

```python
print(s.loc['메로나':'구구콘'])
```

```python
indice = [0, 2]
print(s.iloc[indice])
```

```python
indice = ["메로나", "하겐다즈"]
print(s.loc[indice])
print(s.loc[["메로나", "하겐다즈"]])
```

```python
s
```

```python
s.loc["메로나"] = 500 # s.iloc[0] = 500, s["메로나"] = 500
print(s)
```

```python
s.loc["비비빅"] = 500
print(s)
```

```python
s = s.drop("메로나")
print(s)
```

```python
s = pd.Series([9904312, 3448737, 2890451, 2466052], index=["서울", "부산", "인천", "대구"])
s2 = pd.Series({"서울": 9631482, "부산": 3393191, "인천": 2632035, "대전": 1490158}, index=["부산", "서울", "인천", "대전"])
ds = s - s2
print(ds)
```

```python
s3 = s / 100000
print(s3)
```

```python
print(s.values - s2.values)
```

```python
print(ds[ds.notnull()])
```

```python
print(s)
print()
print(s2)
print()
rs = (s - s2) / s2 * 100
print(rs)
print()
rs = rs[rs.notnull()]
print(rs)
```

```python
rs["부산"] = 1.63
rs["대구"] = 1.41
print(rs)
```

```python
del rs["서울"]
print(rs)
```

```python
from pandas import Series

date = ["6/1", "6/2", "6/3", "6/4", "6/5"]
high = Series([42800, 42700, 42050, 42950, 43000], index=date)
low = Series([42150, 42150, 41300, 42150, 42350], index=date)
print(high)
print('---------------------')
print(low)
```

```python
diff = high - low
diff
```

```python
print(diff.idxmax())
```

```python
def remove_comma(x):
    print(x, 'in function')
    return int(x.replace(",", ""))
```

```python
s = Series(["1,234", "5,678", "9,876"])
result = list(s.map(remove_comma))
print(result)
```

```python
from pandas import Series

data = [42500, 42550, 41800, 42550, 42650]
index = ['2019-05-31', '2019-05-30', '2019-05-29', '2019-05-28', '2019-05-27']
s = Series(data=data, index=index)
print(s)
```

```python
cond = s > 42000
print(cond)
```

```python
print(s[cond])
```

```python
print(s[s > 42000])
```

```python
from pandas import Series

data = [3.1, 2.0, 10.1, 5.1]
index = ['000010', '000020', '000030', '000040']
s = Series(data=data, index=index)
print(s)
```

```python
s1 = s.sort_values(ascending=False)
print(s1)
```

```python
s2 = s.rank(ascending=False)
print(s2)
```

```python
a = Series(dict(zip([1, 2, 3, 4], ["a", "b", "c", "d"])), [2, 4])
print(a)
print()
print(a.index)
print()
print(a.values)
```
