```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)

plt.xticks(np.arange(0, 2.2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.tick_params(axis='x', direction='in', length=3, pad=6, labelsize=14, labelcolor='green', top=True)
plt.tick_params(axis='y', direction='inout', length=10, pad=15, labelsize=12, width=2, color='r')

plt.show()
```

tick_params() 함수 : 눈금의 스타일을 다양하게 설정
- axis: 설정이 적용될 축을 지정
- direction: 'in', 'out'으로 설정하면 눈금이 안/밖으로 표시, {'in', 'out', 'inout'} 중 선택
- length: 눈금의 길이를 지정
- pad: 눈금과 레이블과의 거리를 지정
- labelsize: 레이블의 크기를 지정
- labelcolor: 레이블의 색상을 지정
- top/bottom/left/right: True/False로 지정하면 눈금이 표시될 위치를 선택
- width: 눈금의 너비를 지정
- color: 눈금의 색상을 지정

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

mpl.rcParams['axes.unicode_minus'] = False
x1 = np.linspace(start=0.0, stop=5.0, num=50)
x2 = np.linspace(0.0, 2.0, num=50)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.suptitle("Main Title")
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2-1 subplots')
plt.xlabel('time (m)')
plt.ylabel('Dampled oscillation')
plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.title('A tale of 2-2 subplots')
plt.xlabel('time (s)')
plt.ylabel('Undampled')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)
plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='hot')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 12)
plt.rc('axes', unicode_minus = False)

year = [2018, 2019, 2020, 2021, 2022]
# 만명단위
physics = [6.2, 6.3, 6.0, 5.9, 6.8]
chemistry = [10.6, 9.4, 8.0, 8.0, 8.0]
life = [16.1, 16.3, 13.9, 13.1, 14.7]
earth = [16.7, 17.7, 16.1, 13.0, 14.9]

science = {
    "year": year,
    "physics": physics,
    "chemistry": chemistry,
    "life": life,
    "earth": earth
}

sci = pd.DataFrame(science)
sci
```

```python
plt.figure(figsize = (10, 5))
plt.bar(sci['year'], sci['earth'])
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.bar(sci['year'], sci['earth'])
plt.bar(sci['year'], sci['life'], bottom = sci['earth'])
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.bar(sci['year'], sci['earth'])
plt.bar(sci['year'], sci['life'], bottom = sci['earth'])
plt.bar(sci['year'], sci['chemistry'], bottom = sci['earth'] + sci['life'])
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.title('학년도 기준 수능 과학탐구 응시자 수')
plt.xlabel('학년도')
plt.ylabel('응시자 수(단위: 만명)')
plt.bar(sci['year'], sci['earth'], label='지구과학')
plt.bar(sci['year'], sci['life'], bottom=sci['earth'], label='생명과학')
plt.bar(sci['year'], sci['chemistry'], bottom=sci['earth'] + sci['life'], label='화학')
plt.bar(sci['year'], sci['physics'], bottom=sci['earth'] + sci['life'] + sci['chemistry'], label='물리학')
plt.legend(ncol=2)
plt.show()
```
