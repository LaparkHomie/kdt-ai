```python
import pandas as pd
import numpy as np
```

```python
# 문제 1

s1 = pd.Series(data=[10, 20, 30, 40, 50], index=["가", "나", "다", "라", "마"])
s1
```

```python
# 문제 2

np.random.seed(20)
sample2 = pd.Series(np.random.randint(100, 200, size=(15,)))
sample2 = sample2[sample2 <= 160]
sample2
```

```python
# 문제 3

s3 = pd.Series(data=["apple", np.nan, "banana", "kiwi", "gubong"], index=["가", "나", "다", "라", "마"])
s3
```

```python
# 문제 4

s4 = s3[s3.notna()]
s4
```

```python
# 문제 5

s5 = s3[s3.isna()]
s5
```
