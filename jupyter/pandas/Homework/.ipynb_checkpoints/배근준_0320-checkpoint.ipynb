### [문제] Seaborn이 제공하고 있는 데이터 셋을 사용하여 질문에 답하시오.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore') # 경고무시

plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

### (1) 팁 데이터에서 pivot_table()함수를 사용하여 index='day', columns='size', values='tip'으로하는 데이터를 만든 다음 해당 데이터로 히트맵을 작성하시오.

```python
tips_table = tips.pivot_table(index="day", columns="size", aggfunc="size")

# 결측치 제거
tips_table.fillna(tips_table.mean(), inplace=True)

sns.heatmap(tips_table, cmap=sns.light_palette("pink", as_cmap=True), annot=True)
plt.title("요일별 팁 히트맵", y=1.05, fontsize=20)
plt.show()
```

### 아이리스 데이터에서 각 데이터의 상관관계를 pairplot()으로 나타내시오.

```python
iris_corr = iris.corr(numeric_only = True)

sns.pairplot(iris_corr, diag_kind="kde", markers=["o", "s", "D"], height=3)
plt.suptitle("아이리스 데이터의 상관관계", y=1.05, fontsize=20)
plt.show()
```
