```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 한강

```python
HAN_CSV_PATH = (
    "raw/Han_2019-2020.csv",
    "raw/Han_2021.csv",
    "raw/Han_2022.csv",
    "raw/Han_2023.csv"
)
```

### 금강

```python
GEUM_CSV_PATH = (
    "raw/Geum_2019-2020.csv",
    "raw/Geum_2021-2023.csv"
)
```

### 낙동강

```python
NAKDONG_CSV_PATH = (
    "raw/Nakdong_2019-2020.csv",
    "raw/Nakdong_2021-2022.csv",
    "raw/Nakdong_2023.csv"
)
```

### 영산강섬진강

```python
YEONGSANSEOMJIN_CSV_PATH = (
    "raw/YeongsanSeomjin_2019-2020.csv",
    "raw/YeongsanSeomjin_2021-2023.csv"
)
```

### 한국환경공단 공공하수처리시설 현황

```python
PSTF_CSV_PATH = {
    2019: "raw/Public Sewage Treatment Facility_2019.csv",
    2020: "raw/Public Sewage Treatment Facility_2020.csv",
    2021: "raw/Public Sewage Treatment Facility_2021.csv",
    2022: "raw/Public Sewage Treatment Facility_2022.csv",
    2023: "raw/Public Sewage Treatment Facility_2023.csv"
}
```

```python
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

# 수계별 폐수물 방류량
pstf_lst = []
for year, path in PSTF_CSV_PATH.items():
    df = pd.read_csv(path, encoding="cp949")

    df["Year"] = year
    pstf_lst.append(df)

df = pd.concat(pstf_lst, ignore_index=True)
df_waste = df.groupby(["Year", "수계"])["방류량"].sum().reset_index()

# 수계별 toc 수치
wq_paths = HAN_CSV_PATH + GEUM_CSV_PATH + NAKDONG_CSV_PATH + YEONGSANSEOMJIN_CSV_PATH
wq_lst = []
for path in wq_paths:
    df = pd.read_csv(path, encoding="cp949")
    wq_lst.append(df)

df = pd.concat(wq_lst, ignore_index=True)
df_toc = df[df["항목"] == "TOC"]

date_columns = df_toc.columns[4:] # 0~2 수계, 3 항목 고정, 4부터 날짜

lst_data = []
for col in date_columns:
    df_tmp = df_toc[["수계별(1)", col]]
    df_tmp.columns = ["수계", "TOC"]
    df_tmp["Date"] = col
    lst_data.append(df_tmp)

df = pd.concat(lst_data, ignore_index=True)

df["Year"] = df["Date"].str[:4].astype(int)
#df["Year"] = map(int, df["Date"][:4])
df["TOC"] = pd.to_numeric(df["TOC"], errors='coerce')
df_avg = df.groupby(["Year", "수계"])["TOC"].mean().reset_index()
df_avg.columns = ["Year", "수계", "TOC"]

df_final = pd.merge(df_waste, df_avg, on=["Year", "수계"], how="inner")

plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df_final, 
    x='방류량', 
    y='TOC', 
    hue='Year',
    style='수계',
    s=200, 
    palette='viridis',
    alpha=0.7
)

plt.title("수계별 총 방류량 대비 TOC 수질 현황 (2019-2023)", fontsize=15)
plt.xlabel("총 방류량 합계 (㎥/일)", fontsize=12)
plt.ylabel("TOC 평균 수치 (㎎/ℓ)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

# 범례
plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left", borderaxespad=0)
plt.tight_layout()
plt.show()
```
